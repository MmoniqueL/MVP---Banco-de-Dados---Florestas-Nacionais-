# 4.Modelagem do Banco de Dados

Dado que a conjunto de dados não é excessivamente complexa e considerando que o objetivo é realizar consultas analíticas e relatórios, o esquema relacional proposto foi  o estrela. Ele oferece simplicidade, desempenho superior em consultas comuns e é mais fácil de implementar e manter. Este esquema onsiste em uma tabela de fatos central e na criação de várias tabelas dimensões que se relacionam diretamente com a tabela fatos. Tais tabelas serão  criadas como descrito abaixo:


**Tabela de Fatos**:
FatoFlorestasColunas: id, Tipologia_id, Categoria_id, Bioma_id, Ano, Area

**Dimensões**:
DimTipologia: id, Tipologia_sigla, Tipologia
DimCategoria: id, Categoria, cod_tipo, cod_class
DimBioma: id, Bioma

## 4.2 Tabela de fatos

A tabela de fatos no contexto do banco de dados relacional sobre Florestas Naturais do Brasil contém medidas e métricas centrais do projeto. Abaixo estão detalhados o significado de cada coluna na tabela de fatos `FatoFlorestas`:

### Tabela de Fatos: `FatoFlorestas`

1. **id**
   - **Descrição**: Identificador único para cada registro na tabela de fatos.
   - **Função**: Serve como chave primária para garantir a unicidade de cada entrada na tabela. Facilita operações de gerenciamento de dados, como atualizações e exclusões.

2. **Tipologia_id**
   - **Descrição**: Referência à chave primária da tabela de dimensão `DimTipologia`.
   - **Função**: Conecta a tabela de fatos com a tabela de dimensão que contém informações sobre a tipologia da floresta (tipologia e sigla da tipologia).

3. **Categoria_id**
   - **Descrição**: Referência à chave primária da tabela de dimensão `DimCategoria`.
   - **Função**: Conecta a tabela de fatos com a tabela de dimensão que contém informações sobre a categoria da floresta (categoria, código de tipo e código de classe).

4. **Bioma_id**
   - **Descrição**: Referência à chave primária da tabela de dimensão `DimBioma`.
   - **Função**: Conecta a tabela de fatos com a tabela de dimensão que contém informações sobre o bioma da floresta.

5. **Ano**
   - **Descrição**: O ano específico ao qual o registro se refere.
   - **Função**: Permite a análise temporal dos dados, mostrando como as áreas florestais mudaram ao longo dos anos.

6. **Area**
   - **Descrição**: A área florestal medida para um determinado ano.
   - **Função**: Representa a medida central dos dados na tabela de fatos. É a métrica que será analisada, permitindo a identificação de tendências e padrões de desmatamento ou recuperação de áreas florestais.

## 1. Como as áreas floreiatais evoluíram ao longo dos anos?

O seguinte código SQL foi utilizado afim de  responder a pergunta acima:

```
# SELECT Ano, SUM(Area) AS Total_Area
FROM FatoFlorestas
GROUP BY Ano
ORDER BY Ano;

```

O resultado obtido está disponível no gráfico abaixo:

image.png

Podemos observar que do nosso conjunto de dados a área florestal era significativamente maior em 1990 e reduziu amplamente durante a década de 90 e início dos anos 2000. Durante os anos recentes (apõs 2010) redução da área florestal sse mostrou mais lenta.

## 2. Quais foram os biomas mais afetados pelo desmatamento?

O seguinte código SQL foi utilizado afim de  responder a pergunta acima:

```
# SELECT dB.Bioma, SUM(CASE WHEN Ano = 1990 THEN Area ELSE 0 END) AS Area_1990,
               SUM(CASE WHEN Ano = 2025 THEN Area ELSE 0 END) AS Area_2025,
               SUM(CASE WHEN Ano = 1990 THEN Area ELSE 0 END) - SUM(CASE WHEN Ano = 2025 THEN Area ELSE 0 END) AS Area_Loss
FROM FatoFlorestas f
JOIN DimBioma dB ON f.Bioma_id = dB.id
GROUP BY dB.Bioma
ORDER BY Area_Loss DESC;


```
image.png

O resultado demontrou que o bioma com maior perda territorial nos últimos 34 anos foi o Cerrado, seguid pela Amazônia, Caatinga e Pampa.
Ponto de atenção para a regição dos Pampas, que foi reduzida aproximadamente pela metade.

## 3. Quais foram os anos com maior desmatamento?

O seguinte código SQL foi utilizado afim de  responder a pergunta acima:

```
# WITH AnnualChange AS (
    SELECT Ano, SUM(Area) AS Total_Area
    FROM FatoFlorestas
    GROUP BY Ano
)
SELECT Ano, Total_Area,
       LAG(Total_Area) OVER (ORDER BY Ano) - Total_Area AS Area_Loss
FROM AnnualChange
ORDER BY Area_Loss DESC;


```

image.png

O resultado demonstrou que o ano com maior desmatamento foi 2000. seguido por 2004 e 2002 em terceiro. No ultimos anos foi observada uma redução na perda de áreas de florestas.
